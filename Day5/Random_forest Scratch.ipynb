```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_features=5, n_redundant=0, n_informative=5, n_clusters_per_class=1)
```

```python
df = pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target']=y
print(df.shape)
df.head()
```

```python
def sample_rows(df,percent):
    return df.sample(int(percent*df.shape[0]),replace=True)
```

```python
df1=sample_rows(df,0.2)#20% of data in df1
df2=sample_rows(df,0.3)#30% of data in df2
df3=sample_rows(df,0.4)#40% of data in df3
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

```python
#The no of subsets of data = the no of decision trees a person will need to make
```

```python
from sklearn.tree import DecisionTreeClassifier
clf1=DecisionTreeClassifier()
clf2=DecisionTreeClassifier()
clf3=DecisionTreeClassifier()
#we have trained 3 decision trees
```

```python
clf1.fit(df1.iloc[:,0:5],df1.iloc[:,-1])#all rows, represented by iloc[:,-1]
clf2.fit(df2.iloc[:,0:5],df2.iloc[:,-1])
clf3.fit(df3.iloc[:,0:5],df3.iloc[:,-1])
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(clf1)
```

```python
plot_tree(clf2)
```

```python
plot_tree(clf3)
```
