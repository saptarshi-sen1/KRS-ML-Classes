```python
#import packages
import pandas as pd
import numpy as np
from sklearn. model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
#import data
spam_df = pd.read_csv("spam_filter.csv")
```

```python
#inspect data
spam_df.groupby('Category').describe()
```

```python
#turn ham or spam to numerics
#creating a new column called spam
spam_df['spam'] = spam_df['Category'].apply(lambda x: 1 if x == 'spam' else 0)
```

```python
spam_df
```

```python
#create our tran_test_split
X_train,X_test,y_train,y_test = train_test_split(spam_df.Message, spam_df.spam, test_size = 0.25)# we are trying to do different evaluations
# when we use random state, it will just auto-shuffle
```

```python
#find the word count and store data as numerical matrix
cv = CountVectorizer()#every word is assigned a no. and it is then counted
X_train_count = cv.fit_transform(X_train.values)
```

```python
X_train_count.toarray()
```

```python
#train
model =  MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
#pre-test ham
email_ham = ["hey wanna meet up for the movie?"]
email_ham_count = cv.transform(email_ham)
model.predict(email_ham_count)
#answer is zero which is not spam
```

```python
#pre-test spam
email_spam = ["reward money click"]
email_spam_count = cv.transform(email_spam)
model.predict(email_spam_count)
#answer is one which is spam
```

```python
#test model
X_test_count = cv.transform(X_test)
model.score(X_test_count,y_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
y_pred = model.predict(X_test_count)
```

```python
# Accuracy
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)
```

```python
# Precision
prec = precision_score(y_test, y_pred)
print("Precision:", prec)
```

```python
# Recall
rec = recall_score(y_test, y_pred)
print("Recall:", rec)
```

```python
# F1 Score
# it is used to detect both the recall and the precision of a particular model
f1 = f1_score(y_test, y_pred)
print("F1 Score:", f1)
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", cm)
```
