```python
#Importing Required Libraries

import pandas as pd
import numpy as np
```

```python
#Importing dataset from skikit learn. This is taken from the online dataset

from sklearn.datasets import fetch_california_housing ##skikit learn library
```

```python
dataset=fetch_california_housing()
dataset
##ml models have targets(dependent) and features(independent)==> Supervised learning
```

```python
#Converting it into Dataframe - Independent Variables

df_x = pd.DataFrame(dataset.data)# the value of features are provided. The features are columns that are independent and targets are the independent ones.
df_x
```

```python
#Adding Column names from dataset - Independent/Features

df_x.columns=dataset.feature_names
df_x
```

```python
#Dependent Variable/target

df_y = pd.DataFrame(dataset.target, columns=dataset.target_names)
df_y
```

```python
#Independent features and Dependent features

x=df_x#dependent
y=df_y#independent
#preprocessing is complete till this part
```

```python
#Train Test Split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=42)#it means 70%training and 30% testing; random state is used for shufling
# 70% of x and 70% of y is taken, not seperate parts of individuals are taken.
```

```python
#basically we are pre-processing the data.
```

```python
#Standardizing dataset
# we are doing weighing. It is for the variations between the values. If the values are high, the model assumes that it has higher priority and take/gives its values, we standardizr to increase variation

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
##Scalar = (x-(miu))/sigma
```

```python
#Fitting in Standard Scaler
#fit_transform --> taking te values from the formuales, it takes few values. Transform--> it takes and replaces the values from the standardized value to the initial values
x_train=scaler.fit_transform(x_train)#the training data is taken and 
x_test=scaler.transform(x_test)# we are taking the mean values of train and replace it with the values and then fit it. We are using the same test scale
```

```python
#Model Initialization

from sklearn.linear_model import LinearRegression
model=LinearRegression()#model reinitialization. We have not provided the data to the model 
```

```python
#Fitting my dataset into the model 

model.fit(x_train,y_train)
```

```python
#Prediction

model_pred=model.predict(x_test)#We used x_test as we had kept that seperate for proper prediction as if we give the traiing data for testing it would always give us the correct answer.
model_pred
```

```python
#Evaluation 

from sklearn.metrics import mean_squared_error#mean_squared_error --> it is a built in feature

rmse = np.sqrt(mean_squared_error(y_test, model_pred))#(actual-predicted)/predicted; we can improve its accuracy by dealing with random state(1-100)
rmse
```
